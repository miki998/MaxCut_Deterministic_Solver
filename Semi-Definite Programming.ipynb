## Semi-Definite Programming

An optimization type of programming and part of the approximation algorithms class.

### General Understanding
- Solve $\max \sum_{i,j}c_{ij}x_{ij}$ with $\sum_{i,j} a_{ij}x_{ij}=b_k$ and $X=(x_{ij})$ positive semi-definite symmetric
- Used as a form for vector programming.
- Property of, strong/weak duality and complementary slackness($<x,s>=0$ if $x,s$ optima) not assured 

### Duality of SDP
- Dual form of SDP:
    - $\max <b,y>$  ; $y_1A_1+...+y_mA_m$ positive semi-definite
    - with $A_i$  and $C$ in positive cone of matrix.
- The Lagrangien $L(X;Y;\lambda)=<C,X>-<X,Y>+\sum_{1\geq i\geq m} \lambda_i(b_i-<A_i,X>)$, $X,Y$ positive semi-definite cone
- Solving SDP is similar to solve  $\min_{X\in S_n} \sup_{Y\in S_n^+,\lambda\in R^m} L(X;Y,\lambda)$

### Methods to Solve SDP
- Interior point methods (multiple directions)
- Center Path methods (multiple directions)
- Bundle method
- Gradient Projection

Potential Reduction (e.g Interior points): is to minimize a potential function and not an objective one. It should reduce the complementary slackness, but also should either have a reduction of value for either a primal step or dual step.
- Primal Dual Algorithm: (suited for dense graph but not exploit sparse graph)
    - minimize $<I_n,x>$ with $w+x$ positive semi-definite and $x_{ij}=1_{i=j}$ (similar to max eigenvalues reduction)
    - minimize $<s,w>$ subject to $s$ positive semi-definite and $s_{ii}=1$ (similar to maximizing $1-s$ so same as general stipulation of maxcut)
    - Potential function composed of log barrier for positive cone: $-\ln(\det(u))$, and $\alpha (<s,u>/<s,x>-1)$ (addition of both)
    - $O(\sqrt{n})$ steps later, the potential function is sufficiently small.
