## Linear Programming

An optimization type of programming and part of the approximation algorithms class.

### General understanding

- Solve $\max\sum_{j=1}^nc_jx_j$ with $\sum_{j=1}^na_{ij}x_j\leq b_i$ and $x_j\geq 0$
- $m$ constraints gets us  positivity and $m$ bounds to form a polyhedron.
- Potentially Lemme DE FARKAS

### Dual of a LP

- Given a LP: $\;$  $\max cx \;|\;Ax\leq b, x\geq 0\;$    we associate the following Dual:
$$\min\lambda b \;|\;  \lambda A\geq c, \lambda \geq 0 $$
- Weak duality: $cx\geq\lambda b$ for all feasible solution $x$ and feasible solution $\lambda$.
- Criteria for opimality:
    - $\lambda_i(b_i-\sum_j a_{ij}x_j)=0, \;\;i=1,...,m$
    - $(\sum_i\lambda_ia_{ij}-c_j)x_j=0, \;\;j=1,...,n$
- Strong duality: $c^Tx=b^Ty$
- Advantage of Dual: 
    - bound for primal solution
    - potential faster runtime

### Solving of LP

- If linear function reach max or min then this OPT is reached on an extremum point. 
    - $I(x^0)=\left\{i|\alpha_ix^0=b_i\right\}$
    - $L(x^0)=\left\{x|\alpha_ix=b_i \;\forall i\in I(x^0)\right\}$
    - naming $\alpha_ix$ the $i$-th line of constraint and $x^0$ a feasible vector.
- Solving of system using Gauss-Jordan

#### <u>Simplex algorithm: 
Chapter 5 Matouchek
    
Notion of slack variables (variables added to make the inequality constraints, equalities constraint)

- Basic variables: variables not currently set to zero by simplex
- Non basic variables: variables set to zero by simplex

##### General idea
- Idea : go through extremities of polytope defined above.
- Steps for simplex: 
    - 1. find feasible vertex (extremity) and swap a basic variable to improve if does not exist then stop
    - 2. select variables to swap in and out in greedy fashion
    - 3. update situations then repeat steps
- for m equations (constraints) gives typically between 2m and 3m pivot steps
- Klee-Minty found example with n constraints and $2^n-1$ pivot steps
  
##### Rules for initialization and others
- Apply simplex to auxiliary problem
- many rules to pivot (swap basic variables) (steepest edge/largest coefficient/largest increase/Random edge/Bland's rule)
- Exception Handling: Unboundedness / Degeneracy / Infeasibility
